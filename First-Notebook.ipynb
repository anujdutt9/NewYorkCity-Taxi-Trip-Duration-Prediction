```python
# Import Dependencies
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load Dataset
train_data = pd.read_csv('./Dataset/train.csv')

train_data.head()
```

Find Feature Correlation. Since, <b>id</b> is always unique, so it's not a useful feature. Let's check it's correlation with the labels i.e. <b>trip_duration</b>.

```python
print(train_data['pickup_datetime'][0].split(' ')[1])
```

```python
for i in range(5):
    t1 = train_data['pickup_datetime'][i].split(' ')[1]
    t2 = train_data['dropoff_datetime'][i].split(' ')[1]
    FMT = '%H:%M:%S'
    time_diff = datetime.strptime(t2, FMT) - datetime.strptime(t1, FMT)
    print("Time Difference: ", time_diff)
    print("Time Difference (in seconds): ", time_diff.total_seconds())
```

The <b>vendor_id</b> could be a thing that might define the trip duration. People might love to take one type of provider over other based on rates offered, distance to travel etc.

```python
# vendor_id vs trip_duration
train_data['vendor_id'].corr(train_data['trip_duration'])
```

Pickup date and time might be a god indicator of trip duration as on holidays, there might be a less traffic as compared to when people are at work. Also, people might like to take a ride when they go home or are going to office i.e. in morning or evening during workdays. So, during the day, the number of ride might be less and hence less duration.

```python
# pickup_datetime
train_data['pickup_datetime'].corr(train_data['trip_duration'])
```
